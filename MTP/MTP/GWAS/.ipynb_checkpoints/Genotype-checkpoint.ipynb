```python
import pandas as pd
big_data=pd.read_csv("X.txt" , sep=" " , chunksize=200)
```

```python
import numpy as np
genotype=pd.DataFrame()
for chunk in big_data:
    df=pd.DataFrame(chunk).astype(np.int8)
    genotype=pd.concat([genotype,df], ignore_index=True,) 
```
