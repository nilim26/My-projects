```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
d1=np.random.rand(1000)
for i in range(0,1000):
    if d1[i]>=.5:
        d1[i]=1
    else:
        d1[i]=0
d1
        
```

```python
intd1=[int(x) for x in d1]
print(sum(intd1))                     #to see how many 1's are there
```

```python
phenotype=pd.DataFrame({'Phenotype':intd1})
```
